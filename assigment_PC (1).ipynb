## Exercise 1
Write a simple, short program to calculate and print area of one of the 
biggest faces and volume of a rectangular cuboid with the dimensions 
given below:
Length (l) = 30cm
Breadth (b) = 20cm
Height (h) = 15cm

```python
def cuboid(l,b,h):
    a=[l,b,h]
    a.sort(reverse=True)
    print(' area of one of the biggest faces is ',a[0]*a[1],'dim=',a[0],'*',a[1])
    print('volume of a rectangular cuboid is',a[0]*a[1]*a[2],'dim=',a[0],'*',a[1],'*',a[2])
```

```python
cuboid(30,15,20)
```
